```python
import csv

f=open('incheon_population.csv', 'r', encoding='cp949')
data = csv.reader(f)

result = []
name = '연수1동'

for row in data:
    if str(name) in row[0]: 
        for m, fm in zip(row[3:104], row[106:207]): 
            result.append(int(m)+int(fm))
    
# print(result)
f.close()

plt.rc('font', family='AppleGothic')
plt.title(str(name) + '연령별 인구수')

plt.style.use('ggplot')
plt.bar(range(101), result)
plt.xlabel('age')
plt.ylabel('population')
plt.show()
```

# 우리 동네 항아리 그래프 그리기

```python
f=open('incheon_population.csv', 'r', encoding='cp949')
data = csv.reader(f)

male = []
female = []
name = '연수1동'

for row in data:
    if name in row[0]:
        for i in row[3:104]: 
            male.append(-int(i))
        for i in row[106:207]: 
            female.append(int(i))
            
f.close()

plt.rc('font', family='AppleGothic')
plt.title(str(name) + '연령별 인구수')

# fix minus no show problem
plt.rcParams['axes.unicode_minus'] = False
plt.barh(range(101), male, label='남성')
plt.barh(range(101), female, label='여성')
plt.xlabel('age')
plt.ylabel('population')
plt.legend()
plt.show()
```
