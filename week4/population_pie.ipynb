```python
import matplotlib.pyplot as plt

# 타원이 나오는 경우가 있음. 원으로 나오게 고정.
plt.axis('equal')
# 리스트로 받은 입력을 자동으로 파이차트로 표현해줌
plt.pie([10, 20])
plt.show()
```

```python
plt.rc('font', family='AppleGothic')
# 타원이 나오는 경우가 있음. 원으로 나오게 고정.
plt.axis('equal')
# 리스트로 받은 입력을 자동으로 파이차트로 표현해줌
# autopct. 비율을 숫자로 표시
# explode. 돌출 효과로 강조
plt.pie([10, 20], labels=['label 1', 'label 2'], autopct='%.2f%%', explode=(0, 0.1))
plt.legend()
plt.show()
```

# 우리 동네 인구 구조를 파이차트로 그리기

```python
import csv

f=open('incheon_population.csv', 'r', encoding='cp949')
data = csv.reader(f)

male = []
female = []
result = []
name = '연수1동'

for row in data:
    if name in row[0]:
        for i in row[3:104]: 
            male.append(int(i))
        for i in row[106:207]: 
            female.append(int(i))
            
result.append(sum(male))
result.append(sum(female))
            
f.close()

plt.rc('font', family='AppleGothic')

plt.axis('equal')
label = ['남', '여']
color = ['crimson', 'darkcyan']
# 리스트로 받은 입력을 자동으로 파이차트로 표현해줌
# autopct. 비율을 숫자로 표시
# explode. 돌출 효과로 강조
# startangle. setting angle
plt.pie(result, labels=label, autopct='%.1f%%', colors=color, startangle=90)
plt.legend()
plt.show()

```
