```python
import matplotlib.pyplot as plt
```

```python
import csv

f=open('incheon_population.csv', 'r', encoding='cp949')
data = csv.reader(f)

male = []; female = []; 
name = '연수1동'

for row in data:
    if name in row[0]:
        for i in row[3:104]: 
            male.append(int(i))
        for i in row[106:207]: 
            female.append(int(i))
            
f.close()

plt.rc('font', family='AppleGothic')
plt.title('연령대별 성별 인구 그래프')
plt.scatter(male, female, c=range(101), alpha=0.5, cmap='jet')
plt.colorbar()
line = range(max(male))
plt.plot(line, line, color='g')
plt.xlabel('male')
plt.ylabel('female')
plt.show()
```

```python
import csv
import math

f=open('incheon_population.csv', 'r', encoding='cp949')
data = csv.reader(f)

male = []; female = []; size = []
name = '연수1동'

for row in data:
    if name in row[0]:
        for i in row[3:104]: 
            male.append(int(i))
        for i in row[106:207]: 
            female.append(int(i))
            
for i in range(len(male)):
    size.append(math.sqrt(male[i]+female[i]))
f.close()

plt.rc('font', family='AppleGothic')
plt.title('연령대별 성별 인구 그래프')
plt.scatter(male, female, s=size, c=range(101), alpha=0.9, cmap='jet')
plt.colorbar()
line = range(max(male))
plt.plot(line, line, color='g')
plt.xlabel('male')
plt.ylabel('female')
plt.show()
```
