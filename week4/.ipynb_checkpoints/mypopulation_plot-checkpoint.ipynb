```python
f=open('incheon_population.csv', 'r', encoding='cp949')
data = csv.reader(f)

male = []

for row in data:
    if '송도1동' in row[0]: 
        for i in row[3:104]: 
#             print(i, end=' ')
            male.append(int(i))
    
print(male)
f.close()
```

```python
f=open('incheon_population.csv', 'r', encoding='cp949')
data = csv.reader(f)

result = []
name = input('인구 구조를 알고 싶은 동 이름 입력: ')

for row in data:
    if str(name) in row[0]: 
        for m, fm in zip(row[3:104], row[106:207]): 
            result.append(int(m)+int(fm))
    
print(result)
f.close()

plt.rc('font', family='AppleGothic')
plt.title(str(name) + '연령별 인구수')

plt.style.use('ggplot')
plt.plot(result)
plt.xlabel('age')
plt.ylabel('population')
plt.show()
```

# 실습: 내가 사는 동네의 인구 구조를 시각화해보고, 시각화 결과를 분석해보기
저는 인천광역시 연수구 연수1동에 살고 있습니다.


## 인천광역시 연수구 연수1동의 인구 구조 시각화

```python
f=open('incheon_population.csv', 'r', encoding='cp949')
data = csv.reader(f)

result = []
name = '연수1동'

for row in data:
    if str(name) in row[0]: 
        for m, fm in zip(row[3:104], row[106:207]): 
            result.append(int(m)+int(fm))
    
# print(result)
f.close()

plt.rc('font', family='AppleGothic')
plt.title(str(name) + '연령별 인구수')

plt.style.use('ggplot')
plt.plot(result)
plt.xlabel('age')
plt.ylabel('population')
plt.show()
```

## **연수1동** 인구 구조 시각화로 알 수 있는 사실들

- 50대 후반이 가장 많이 살고 있습니다
- 50대 초반부터 60대 초반까지 인구가 많습니다
- 20대 중반부터 30대 초반까지 인구가 많습니다
- 30대 인구는 20대, 40대, 50대, 60대 인구에 비해 적습니다
- 70대 이상은 많지 않습니다
- 청소년 인구가 많지 않습니다
