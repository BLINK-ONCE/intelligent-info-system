```python
import matplotlib.pyplot as plt
import csv
```

```python
f = open('incheon_weather.csv', 'r')
bday = '2001-02-16'
years = list()
low_temps = list()
high_temps = list()
data = csv.reader(f)
header = next(data)

count = 0
print(header)
for row in data:
    if len(row) == 5: date, loc, avg_temp, low_temp, high_temp = row
    if low_temp == '' or high_temp =='': continue
#     print(date[8:10])
    if int(bday[0:4]) <= int(date[0:4]) and int(bday[5:7]) == int(date[5:7]) and int(bday[8:10]) == int(date[8:10]):
        years.append(int(date[0:4]))
        low_temps.append(float(low_temp))
        high_temps.append(float(high_temp))

f.close()
```

```python
import matplotlib
matplotlib.rcParams['axes.unicode_minus'] = False
plt.rc('font', family='AppleGothic')
plt.title(bday[0:4] + '년부터 ' + bday[5:7] + '월' + bday[8:10] + '일의 최저 기온과 최고 기온')
plt.plot(years, low_temps, color='b', label='최저 기온')
plt.plot(years, high_temps, color='r', label='최고 기온')
plt.xlabel('연도')
plt.ylabel('°C')
plt.legend()
plt.show()
```
