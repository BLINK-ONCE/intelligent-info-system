````python
from bs4 import BeautifulSoup
from urllib.request import urlopen

url = 'https://www.naver.com'
html = urlopen(url)

soup = BeautifulSoup(html.read(), 'html.parser')

# class가 service_area 인 div를 추출하여 result 에 저장
result = soup.find('div', {'class': 'service_area'})
# result.find('a') 명령어를 통해 첫 번째에 나오는 a태그를 찾은 후 a_result에 저장
a_result = result.find('a')
# a_result.text 를 통해 a 태그 안에 있는 text 만 추출
print(a_result.text)
````

````python
from bs4 import BeautifulSoup
from urllib.request import urlopen

url = 'https://www.naver.com'
html = urlopen(url)

soup = BeautifulSoup(html.read(), 'html.parser')

# <a> 태그와 id 활용하기
result = soup.find('a', {'id': 'NM_set_home_btn'})
print(result.text)
````
