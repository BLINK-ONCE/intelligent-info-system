# 1. pandas.date_range()의 기간을 변경해보기

```python
%%time

from urllib.request import urlopen
from bs4 import BeautifulSoup
from tqdm.notebook import tqdm

date_list = pd.date_range('2022-07-01', periods=100, freq='D')

movie_title = []
movie_point = []
movie_date = []

for today in tqdm(date_list):
    url = 'https://movie.naver.com/movie/sdb/rank/rmovie.naver?sel=pnt&date=' + (today.strftime('%Y%m%d'))
    html = urlopen(url)
    
    soup = BeautifulSoup(html.read(), 'html.parser')
    
    titles = soup.find_all('div', {'class': 'tit5'})
    points = soup.find_all('td', {'class': 'point'})
    
    titles_text = []
    points_text = []
    date = []
    for t, p in zip(titles, points):
        titles_text.append(t.text.strip(' \n'))
        points_text.append(float(p.text))
        date.append(today)
        
    movie_title += titles_text
    movie_point += points_text
    movie_date += date
    
print(movie_title[:5])
print(movie_point[:5])
print(movie_date)
```

## 해당 기간을 설정한 이유
지난 주말 영화관에서 '극장판 짱구는 못말려 : 수수께끼! 꽃피는 천하떡잎학교'를 관람했습니다.

제가 영화관에 가지 못한 최근 4달 동안에는 어떤 영화가 인기였는지 알고 이를 OTT 서비스를 통해 관람하고자 합니다.

# 2. pandas.pivot_table을 이용하여 해당 기간 동안 평균 평점이 가장 높았던 영화 찾기

```python
import pandas as pd

frame = pd.DataFrame({'date': movie_date, 'title': movie_title, 'point': movie_point})
frame
```

```python
import numpy as np


movie_point = pd.pivot_table(frame, values=['point'], index=['title'], aggfunc=np.mean)
movie_point_sort = movie_point.sort_values(by='point', ascending=False)
movie_point_sort.head(10)
```

## 평균 평점이 가장 높았던 영화는 무엇이었으며 이에 대한 자신의 생각
한동안 굉장한 인기를 끌었던 영화, '탑건: 매버릭'이 해당 기간 동안 평균 평점이 가장 높았습니다.

각종 포털 사이트와 소셜 서비스에서 큰 흥행을 기록한만큼 영화 자체의 완성도가 매우 훌륭해 평점 1위를 했을 것이라 생각합니다.
