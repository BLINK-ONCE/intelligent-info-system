# 동계 올림픽 메달 개수 시각화하기
- 동계올림픽 메달 집계 데이터를 참가 횟수와 금메달 개수로 정렬하여 상위 15개국을 막대 그래프로 시각화하기
    - 참가 횟수, 금메달 개수 내림차순 정렬

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = 'https://ko.wikipedia.org/wiki/%EC%98%AC%EB%A6%BC%ED%94%BD_%EB%A9%94%EB%8B%AC_%EC%A7%91%EA%B3%84'

df = pd.read_html(url, header = 0, index_col = 0)
```

```python
data = df[0].iloc[:, 5:10]
data.columns = ['동계 참가 횟수', '금', '은', '동', '계']
data = data.sort_values(by=['동계 참가 횟수', '금'], ascending=False)

plt.rc('font', family='AppleGothic')
plt.style.use('ggplot')
data[1:16].plot(kind='bar')# 상위 15개국(index0: total)
```
